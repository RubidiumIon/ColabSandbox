```python
import random
def random_hex(start : int, stop : int) -> int:
  '''
  Generate a random integer and return it's hexadecimal representation
  '''
  random_int = random.randint(start, stop)
  return hex(random_int)
```

```python
random_hex(1,1000)
```

```python
def unique_random_samples():
  random_set = set()
  new_count = len(random_set)
  while True:
    old_count = new_count
    new_value = random_hex(1,10)
    random_set.add(new_value)
    new_count = len(random_set)
    if (old_count == new_count):
      print(random_set)
      return random_set

```

```python
random.seed(189)
timeit.timeit(unique_random_samples)
```
